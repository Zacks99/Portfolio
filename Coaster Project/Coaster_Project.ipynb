### Step 4 . Data Preparation 

```python
df=df[['coaster_name',
    # 'Length', 'Speed', 
    'Location', 'Status', 
    #'Opening date',
       #'Type',
         'Manufacturer', 
         #'Height restriction', 'Model', 'Height',
       #'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
       #'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
       #'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
       #'Track layout', 'Fastrack available', 'Soft opening date.1',
       #'Closing date', 'Opened', 'Replaced by', 'Website',
       #'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
       #'Single rider line available', 'Restraint Style',
       #'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced', 'latitude', 'longitude', 'Type_Main',
       'opening_date_clean',
        # 'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph',
        # 'height_value', 'height_unit', 
        'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
```

```python
df.head()
```

```python
df.dtypes
```

```python
# change the type of the opening_date_clean
df['opening_date_clean']=pd.to_datetime(df['opening_date_clean'])
```

```python
#rename our columns 
df=df.rename(columns={'coaster_name':'Coaster_Name',
                    'Location':'Location',
                    'Status':'Status',
                     'Manufacturer':'Manufacturer',
                     'year_introduced':'Year_Introduced',
                    'latitude':'Latitude',
                    'longitude':'longitude',
                    'Type_Main':'Type_Main',
                    'opening_date_clean':'Opening_Date',
                    'speed_mph':'Speed_Mph',
                     'height_ft':'Height_ft',
                    'Inversions_clean':'Inversions',
                    'Gforce_clean':'Gforce'}
          )
```

```python
# see the missing values 
df.isna().sum() 
```

```python
# to se the duplicated values
df.loc[df.duplicated()]
```

```python
# see the duplicated name of coaster 
df.duplicated('Coaster_Name') # we fund some duplicated name 
```

```python
# see the localization of the duplicated names 
df.loc[df.duplicated('Coaster_Name')] # we fund 97 duplicated name  
```

```python
# see one of the duplicated if its identical
df.query('Coaster_Name == "Crystal Beach Cyclone"') 
```

```python
#remove certain duplicate 
df=df.loc[~df.duplicated(subset=['Coaster_Name','Location','Opening_Date'])]
df.reset_index(drop=True).copy()
```

### Step 5 . Feature Understanding  

```python
#Top Year Coaster Introduced
ax=df['Year_Introduced'].value_counts().head(10).plot(kind='bar',title='Top Year Coaster Introduced')
plt.xlabel('Years')
plt.ylabel('Count')
```

```python
# Cost speed (mhp)
ax=df['Speed_Mph'].plot(kind='hist',bins =20, title='Coester speed',)
plt.xlabel('Speed MPH')
plt.ylabel('Frequency')
```

### Step 5 . Feature Relationships  

```python
#Relationship btw Speed and Height
df.plot(kind='scatter',x='Speed_Mph',y='Height_ft',title='Coester Speed vs Height')
plt.show()
```

```python
#using seaborn library 
sns.scatterplot(x='Speed_Mph',y='Height_ft',hue='Year_Introduced',data=df)
plt.show()
```

```python
#comparing multiple Featue
sns.pairplot(df,vars=['Year_Introduced','Speed_Mph','Height_ft','Inversions','Gforce'],hue='Type_Main'
             )
plt.show()
```

```python
#see the correlationes
df_corr=df[['Year_Introduced','Speed_Mph','Height_ft','Inversions','Gforce']].dropna().corr()
```

```python
sns.heatmap(df_corr,annot=True)
```

### Step 6 .what are a locationes with fastest roller coesters (min of 10)

```python
ax=df.query('Location !="Other"').groupby('Location')['Speed_Mph']\
.agg(['mean','count'])\
.query('count>=10')\
.sort_values('mean')['mean']\
.plot(kind='barh',figsize=(12,5),title='Average Coast Speed By Location')
ax.set_xlabel('Average Coast Speed')
plt.show()
```
